```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from decimal import Decimal
```

```python
#apple data
appleData = pd.read_csv('SALESFORCE.csv')
appleData
```

```python
appleData.set_index("Date")
```

```python
#pt2 starts on January 5th due to market being closed on Janurary 3rd and Janurary 4th
appleData1= appleData[appleData["Date"].between('2009-01-01','2011-12-31')]
appleData2= appleData[appleData["Date"].between('2009-01-05','2012-01-01')]
appleData= appleData[appleData["Date"].between('2009-01-01','2012-12-31')]
currentAppleDataPrice= appleData[appleData["Date"].between('2023-03-13','2023-03-14')]
```

```python
currentAppleDataPrice["Close"]
```

```python
appleData1
appleData1.reset_index(inplace = True)
appleData1
```

```python
appleData2
appleData2.reset_index(inplace = True)
appleData2
```

```python
applefinal = pd.merge(appleData1,appleData2,left_index=True,right_index=True)
applefinal
```

```python
applefinal["Daily Net Change Percentage"] = applefinal["Close_y"]/applefinal["Close_x"]
applefinal
```

```python
#156 being the current stock price as of 3/17/2023 of salesforce
applefinal["Daily Net Change Percentage"][0:1] = 184
applefinal
```

```python
applePercentageChange = applefinal["Daily Net Change Percentage"]
applePercentageChange
```

```python
applefinal["Inflation Adjusted"] = applePercentageChange.cumprod(axis=0)
print(applefinal)
```
